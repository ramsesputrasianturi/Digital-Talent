```python
import random
import operator

nilaiTugas = []

def tampilGanjil(nilai):
    nilaiGanjil = []
    for x in nilai:
        if x % 2 != 0:
            nilaiGanjil.append(x)
    
    return sorted(nilaiGanjil)

def lebihDari(nilai):
    nilaiLebih = []
    
    for x in nilai:
        if x > 70:
            nilaiLebih.append(x)
        
    return len(nilaiLebih)

def standarDeviasi(nilai):
    jumlah_data = len(nilai)
    jumlah_data_dikurangi = len(nilai) - 1
    mean_1 = sum(nilai)
    mean_2 = []
    
    for x in nilai:
        mean_2.append(x**2)
        
    mean_1_pangkat = sum(mean_2)
    
    deviasi = (jumlah_data) * (mean_1_pangkat) - (mean_1**2)
    total_deviasi = deviasi / (jumlah_data * jumlah_data_dikurangi)
    
    return deviasi    

def maximum(nilai):
    return max(nilai)

def minimum(nilai):
    return min(nilai)

def modus(nilai):
    tampung = []
    simpan = {}
    simpan2 = {}
    
    for x in nilai:
        count = 0
        for z in nilai:
            if x == z:
                count += 1
        
        simpan[x] = count
        
        
        simpan2 = {
          max(simpan.items(), key=operator.itemgetter(1))[0] : max(simpan.items(), key=operator.itemgetter(1))[1]  
        }
        
    return simpan2
        

for i in range(250):
    nilaiTugas.append(random.randrange(1, 100, 1))
    
    
    
print("Nilai Ganjil : " , tampilGanjil(nilaiTugas))
print("\n")
print("Nilai Yang Lebih dari 70 : " , lebihDari(nilaiTugas))
print("\n")
print("Nilai Standar Deviasi : " , standarDeviasi(nilaiTugas))
print("\n")
print("Nilai Maximum : " ,maximum(nilaiTugas))
print("\n")
print("Nilai Minimum : ", minimum(nilaiTugas))

getKeys = list(modus(nilaiTugas).keys())[0]
getValues = list(modus(nilaiTugas).values())[0]

print("\n")
print("Nilai Modusnya : " , getKeys , ", Tampil sebanyak kali : " , getValues)




    

```
