```python
import pymysql
import pandas as pd

host = 'localhost'
port = '3306'
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
            host= host,
            port=int(port),
            passwd=password,
            db=database,
            charset='utf8'
        )

def query(sql):
    query = pd.read_sql_query(sql, conn)
    return query
```

# Report the account representative for each customer.

```python
query("""
    SELECT * FROM customers a 
    INNER JOIN employees b ON a.salesRepEmployeeNumber = b.employeeNumber
""")
```

# Report total payments for Atelier graphique.

```python
query("""
    SELECT SUM(b.amount) AS total
    FROM customers a
    INNER JOIN payments b ON a.customerNumber = b.customerNumber
    WHERE a.customerName = 'Atelier graphique'
""")
```

# Report the total payments by date

```python
query("""
    SELECT a.customerName, b.paymentDate, SUM(b.amount) AS total FROM customers a
    INNER JOIN payments b USING(customerNumber)
    GROUP BY b.paymentDate
""")
```

# Report the products that have not been sold.

```python
query("""
    SELECT * FROM products a
    LEFT JOIN orderdetails b ON a.productCode = b.productCode
    WHERE b.productCode IS NULL
""")
```

# List the amount paid by each customer.

```python
query("""
    SELECT a.customerName, b.paymentDate, SUM(amount) AS jumlah_yang_dibayarkan
    FROM customers a
    INNER JOIN payments b ON a.customerNumber = b.customerNumber
    GROUP BY b.paymentDate
""")
```

# Who are the employees in Boston?

```python
query("""
    SELECT 
        a.employeeNumber, CONCAT(a.firstname, ' ', a.lastName) AS nama_lengkap, b.city 
    FROM employees a
    INNER JOIN offices b ON a.officeCode = b.officeCode
    WHERE b.city LIKE '%Boston%'
""")
```

# Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.

```python
query("""
    SELECT b.customerNumber, b.customerName, SUM(a.amount) AS total_pembayaran
    FROM payments a
    INNER JOIN customers b ON a.customerNumber = b.customerNumber
    GROUP BY a.customerNumber
    HAVING total_pembayaran > 100000
    ORDER BY total_pembayaran DESC
""")
```

# Report the number of orders 'On Hold' for each customer.

```python
query("""
    SELECT COUNT(*) AS total_hold, a.orderNumber, a.orderDate, a.status, b.customerName
    FROM orders a 
    INNER JOIN customers b ON a.customerNumber = b.customerNumber
    WHERE a.status LIKE '%On Hold%'
    GROUP BY a.customerNumber
""")
```

# List products sold by order date.

```python
query("""
    SELECT c.orderDate, a.*
    FROM products a
    INNER JOIN orderdetails b ON a.productCode = b.productCode
    INNER JOIN orders c ON b.orderNumber = c.orderNumber
    GROUP BY c.orderDate, a.productCode
""")
```

# List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
query("""
    SELECT c.orderDate, a.*
    FROM products a
    INNER JOIN orderdetails b ON a.productCode = b.productCode
    INNER JOIN orders c ON b.orderNumber = c.orderNumber
    WHERE a.productName LIKE '%1940 Ford Pickup Truck%'
    GROUP BY c.orderDate DESC, a.productCode
""")
```

# List the names of products sold at less than 80% of the MSRP.

```python
query("""
    SELECT a.productCode, a.MSRP, a.buyPrice, ((a.MSRP - a.buyPrice) / a.buyPrice * 100) AS persen, a.productName
    FROM products a
    INNER JOIN orderdetails b ON a.productCode = b.productCode
    INNER JOIN orders c ON b.orderNumber = c.orderNumber
    GROUP BY b.productCode
    HAVING persen > 80
    ORDER BY persen DESC
""")
```

# Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
query("""
    SELECT a.productCode, a.MSRP, a.buyPrice, ((a.MSRP - a.buyPrice) / a.buyPrice * 100) AS persen, a.productName
    FROM products a
    INNER JOIN orderdetails b ON a.productCode = b.productCode
    INNER JOIN orders c ON b.orderNumber = c.orderNumber
    GROUP BY b.productCode
    HAVING persen > 100
    ORDER BY persen DESC
""")
```

# List the products ordered on a Monday.

```python
query("""
    SELECT a.orderDate, c.productCode, c.productName
    FROM orders a
    INNER JOIN orderdetails b ON a.orderNumber = b.orderNumber
    INNER JOIN products c ON b.productCode = c.productCode
    WHERE DAYNAME(a.orderDate) = 'Monday'
""")
```

# What is the quantity on hand for products listed on 'On Hold' orders?

```python
query("""
    SELECT a.orderDate, a.status, c.productName
    FROM orders a
    INNER JOIN orderdetails b ON a.orderNumber = b.orderNumber
    INNER JOIN products c ON b.productCode = c.productCode
    WHERE a.status = 'On Hold'
    GROUP BY a.orderDate, c.productCode
""")
```
