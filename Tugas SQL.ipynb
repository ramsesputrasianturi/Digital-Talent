```python
import pymysql
import pandas as pd

host = 'localhost'
port = '3306'
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
            host= host,
            port=int(port),
            passwd=password,
            db=database,
            charset='utf8'
        )

def query(sql):
    query = pd.read_sql_query(sql, conn)
    return query
```

Prepare a list of offices sorted by country, state, city.

```python
query(""" 
    SELECT * FROM offices
    ORDER BY country, state, city
""")
```

How many employees are there in the company?

```python
query("""
    SELECT COUNT(*) AS jumlah_karyawan FROM employees
""")
```

What is the total of payments received?

```python
query("""
    SELECT SUM(amount) AS total_payments FROM payments
""")
```

List the product lines that contain 'Cars'.

```python
query("""
    SELECT * FROM products WHERE productLine LIKE '%Cars%'
""")
```

Report total payments for October 28, 2004.

```python
query("""
    SELECT SUM(amount) AS total_payment FROM payments WHERE paymentDate = '2004-10-28'
""")
```

Report those payments greater than \$100,000.

```python
query("""
    SELECT * FROM payments WHERE amount > 100000
""")
```

List the products in each product line.

```python
query("""
    SELECT productLine, productName FROM products
    ORDER BY productLine, productName
""")
```

How many products in each product line?

```python
query("""
    SELECT productLine, COUNT(*) AS jumlah_produk FROM products
    GROUP BY productLine
""")
```

What is the minimum payment received?

```python
query("""
    SELECT MIN(amount) AS min_payment FROM payments
""")
```

List all payments greater than twice the average payment.

```python
query("""
    SELECT * FROM payments 
    WHERE amount>(SELECT AVG(amount) FROM payments)*2
""")
```

What is the average percentage markup of the MSRP on buyPrice?

```python
query("""
    SELECT MSRP, buyPrice, AVG((MSRP - buyPrice) / buyPrice * 100) AS persen, productName FROM products
""")
```

Report the name and city of customers who don't have sales representatives?

```python
query("""
    SELECT customerName, city FROM customers WHERE salesRepEmployeeNumber IS NOT NULL
""")
```

What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
query("""
    SELECT CONCAT(firstName, ' ', lastName) AS nama_lengkap FROM employees
    WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'
""")
```

Which orders have a value greater than $5,000?

```python
query("""
    SELECT orderNumber, SUM(quantityOrdered * priceEach) as nilai FROM orderdetails
    GROUP BY orderNumber
    HAVING nilai > 5000
    ORDER BY nilai DESC
""")
```
